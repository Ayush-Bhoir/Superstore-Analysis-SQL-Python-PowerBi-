# Visuals

```python
# 1) What are the top5 items with the highest average daily sales?

daily_sales = df.groupby(['Product_Name', 'Order_Date'])['Sales'].sum().reset_index()

avg_daily_sales = daily_sales.groupby('Product_Name')['Sales'].mean().sort_values(ascending=False)

top5_avg_daily_sales = avg_daily_sales.head(5).reset_index()

plt.figure(figsize=(10, 6))
ab = sns.barplot(data=top5_avg_daily_sales, x='Sales', y='Product_Name', palette='viridis')
plt.title('Top 5 Products by Average Daily Sales')
plt.xlabel('Average Daily Sales')
plt.ylabel('Product Name')
plt.tight_layout()

for bar in ab.containers :
    ab.bar_label(bar)

plt.show()
```

```python
# 4) What is the maximum run of days for which total daily sales kept increasing?

daily_total_sales = df.groupby('Order_Date')['Sales'].sum().reset_index()

daily_total_sales = daily_total_sales.sort_values(by='Order_Date')

max_run = 0
current_run = 0
for i in range(1, len(daily_total_sales)):
    if daily_total_sales.iloc[i]['Sales'] > daily_total_sales.iloc[i - 1]['Sales']:
        current_run += 1
        max_run = max(max_run, current_run)
    else:
        current_run = 0  

print(f"Maximum number of consecutive increasing sales days: {max_run + 1} days")

plt.figure(figsize=(12, 5))
sns.lineplot(data=daily_total_sales, x='Order_Date', y='Sales', color='blue')
plt.title('Total Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
